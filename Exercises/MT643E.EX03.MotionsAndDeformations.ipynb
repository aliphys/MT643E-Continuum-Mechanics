## Problem 1

Prove

(a) $A_s = \lambda a_i \frac{\partial X_s}{\partial x_i}$

(b) $\lambda^{-2} = a_i a_j \frac{\partial X_s}{\partial x_i} \frac{\partial X_s}{\partial x_j}$

## Solution to Problem 1

### Part (a): Prove $A_s = \lambda a_i \frac{\partial X_s}{\partial x_i}$

**Proof:**

Consider a material line element $dX_s$ in the reference configuration that transforms to $dx_i$ in the current configuration through the deformation mapping $\mathbf{x} = \mathbf{x}(\mathbf{X})$.

The deformation gradient relates these through:
$$dx_i = \frac{\partial x_i}{\partial X_s} dX_s$$

For the inverse transformation $\mathbf{X} = \mathbf{X}(\mathbf{x})$:
$$dX_s = \frac{\partial X_s}{\partial x_i} dx_i$$

Let $\mathbf{a}$ be a unit vector in the current configuration and $\mathbf{A}$ be the corresponding vector in the reference configuration. If $d\mathbf{x} = \lambda d\mathbf{a}$ (where $\lambda$ is the stretch), then:

$$dx_i = \lambda a_i$$

Substituting into the inverse transformation:
$$dX_s = \frac{\partial X_s}{\partial x_i} dx_i = \frac{\partial X_s}{\partial x_i} \lambda a_i$$

Since $d\mathbf{X} = |\mathbf{A}| d\mathbf{A}$ where $|\mathbf{A}|$ is the magnitude of $\mathbf{A}$, we have:
$$|\mathbf{A}| A_s = \lambda a_i \frac{\partial X_s}{\partial x_i}$$

For unit vectors in the reference configuration ($|\mathbf{A}| = 1$):
$$A_s = \lambda a_i \frac{\partial X_s}{\partial x_i}$$

### Part (b): Prove $\lambda^{-2} = a_i a_j \frac{\partial X_s}{\partial x_i} \frac{\partial X_s}{\partial x_j}$

**Proof:**

From part (a), we have:
$$A_s = \lambda a_i \frac{\partial X_s}{\partial x_i}$$

Taking the dot product of both sides with themselves:
$$A_s A_s = \lambda^2 a_i a_j \frac{\partial X_s}{\partial x_i} \frac{\partial X_s}{\partial x_j}$$

Since $\mathbf{A}$ is a unit vector in the reference configuration:
$$\sum_s A_s A_s = |\mathbf{A}|^2 = 1$$

For a single component $A_s$:
$$A_s^2 = \lambda^2 a_i a_j \frac{\partial X_s}{\partial x_i} \frac{\partial X_s}{\partial x_j}$$

Since $\mathbf{a}$ is also a unit vector ($\sum_i a_i^2 = 1$), and considering the relationship between the material and spatial metrics, we get:

$$1 = \lambda^2 a_i a_j \frac{\partial X_s}{\partial x_i} \frac{\partial X_s}{\partial x_j}$$

Therefore:
$$\lambda^{-2} = a_i a_j \frac{\partial X_s}{\partial x_i} \frac{\partial X_s}{\partial x_j}$$

This relationship shows that the inverse square of the stretch ratio equals the metric tensor component in the reference configuration when projected along the spatial direction $\mathbf{a}$.

## Problem 2

A body undergoes the homogeneous deformation
$$x_1 = \sqrt{2}X_1 + \frac{3}{4} \sqrt{2}X_2$$
$$x_2 = -X_1 + \frac{3}{4}X_2 + \frac{1}{4}\sqrt{2}X_3$$
$$x_3 = X_1 - \frac{3}{4}X_2 + \frac{1}{4} \sqrt{2} X_3$$

Find 
(a) the direction after the deformation of a line element with direction ratios 1:1:1 in the reference configuration
(b) the sketch of this line element

## Solution to Problem 2

### Part (a): Finding the deformed direction

**Step 1: Calculate the deformation gradient F**

The deformation gradient is:
$$F_{ij} = \frac{\partial x_i}{\partial X_j}$$

From the given deformation:
$$\mathbf{F} = \begin{bmatrix}
\sqrt{2} & \frac{3\sqrt{2}}{4} & 0 \\
-1 & \frac{3}{4} & \frac{\sqrt{2}}{4} \\
1 & -\frac{3}{4} & \frac{\sqrt{2}}{4}
\end{bmatrix}$$

**Step 2: Apply F to the initial direction vector**

Initial direction vector with ratios 1:1:1 is:
$$\mathbf{N} = \frac{1}{\sqrt{3}}\begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}$$

The deformed direction (before normalization) is:
$$\mathbf{F} \cdot \mathbf{N} = \begin{bmatrix}
\sqrt{2} & \frac{3\sqrt{2}}{4} & 0 \\
-1 & \frac{3}{4} & \frac{\sqrt{2}}{4} \\
1 & -\frac{3}{4} & \frac{\sqrt{2}}{4}
\end{bmatrix} \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}$$

$$= \begin{bmatrix}
\sqrt{2} + \frac{3\sqrt{2}}{4} \\
-1 + \frac{3}{4} + \frac{\sqrt{2}}{4} \\
1 - \frac{3}{4} + \frac{\sqrt{2}}{4}
\end{bmatrix} = \begin{bmatrix}
\frac{7\sqrt{2}}{4} \\
-\frac{1}{4} + \frac{\sqrt{2}}{4} \\
\frac{1}{4} + \frac{\sqrt{2}}{4}
\end{bmatrix}$$

**Step 3: Calculate direction ratios**

The components are:
- $x_1$: $\frac{7\sqrt{2}}{4}$
- $x_2$: $\frac{\sqrt{2} - 1}{4}$  
- $x_3$: $\frac{\sqrt{2} + 1}{4}$

The direction ratios are:
$$\frac{7\sqrt{2}}{4} : \frac{\sqrt{2} - 1}{4} : \frac{\sqrt{2} + 1}{4}$$

Multiplying by 4 to simplify:
$$7\sqrt{2} : (\sqrt{2} - 1) : (\sqrt{2} + 1)$$

**Numerical values:**
$$9.899 : 0.414 : 2.414$$

Or approximately: **9.9 : 0.4 : 2.4**

```python
# Numerical calculation for Problem 2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the deformation gradient F
F = np.array([
    [np.sqrt(2), 3*np.sqrt(2)/4, 0],
    [-1, 3/4, np.sqrt(2)/4],
    [1, -3/4, np.sqrt(2)/4]
])

print("Deformation gradient F:")
print(F)
print()

# Initial direction vector (1:1:1)
N_initial = np.array([1, 1, 1]) / np.sqrt(3)
print("Initial unit direction vector:")
print(N_initial)
print()

# Apply deformation
deformed_direction = F @ N_initial
print("Deformed direction (before normalization):")
print(deformed_direction)
print()

# Calculate direction ratios
print("Direction ratios:")
print(f"{deformed_direction[0]:.3f} : {deformed_direction[1]:.3f} : {deformed_direction[2]:.3f}")
print()

# Simplify ratios by finding common factor
# Multiply by 4 to clear fractions
simplified_ratios = deformed_direction * 4
print("Simplified direction ratios (×4):")
print(f"{simplified_ratios[0]:.3f} : {simplified_ratios[1]:.3f} : {simplified_ratios[2]:.3f}")
print()

# Express in terms of √2
print("Exact direction ratios:")
print(f"7√2 : (√2 - 1) : (√2 + 1)")
print(f"≈ {7*np.sqrt(2):.3f} : {np.sqrt(2)-1:.3f} : {np.sqrt(2)+1:.3f}")
```

```python
# Part (b): Sketch of the line element transformation
fig = plt.figure(figsize=(12, 5))

# Original configuration
ax1 = fig.add_subplot(121, projection='3d')
origin = np.array([0, 0, 0])
initial_direction = np.array([1, 1, 1])

# Plot initial line element
ax1.quiver(origin[0], origin[1], origin[2], 
          initial_direction[0], initial_direction[1], initial_direction[2],
          color='blue', linewidth=3, label='Initial (1:1:1)')

ax1.set_xlim(0, 2)
ax1.set_ylim(0, 2)
ax1.set_zlim(0, 2)
ax1.set_xlabel('X₁')
ax1.set_ylabel('X₂')
ax1.set_zlabel('X₃')
ax1.set_title('Reference Configuration')
ax1.legend()

# Deformed configuration
ax2 = fig.add_subplot(122, projection='3d')

# Calculate deformed direction (scaled for visualization)
deformed_direction = F @ initial_direction
scale_factor = 1.5  # Scale for better visualization

ax2.quiver(origin[0], origin[1], origin[2], 
          deformed_direction[0], deformed_direction[1], deformed_direction[2],
          color='red', linewidth=3, label='Deformed')

# Set appropriate limits
max_val = max(abs(deformed_direction)) * 1.2
ax2.set_xlim(-max_val, max_val)
ax2.set_ylim(-max_val, max_val)
ax2.set_zlim(-max_val, max_val)
ax2.set_xlabel('x₁')
ax2.set_ylabel('x₂')
ax2.set_zlabel('x₃')
ax2.set_title('Current Configuration')
ax2.legend()

plt.tight_layout()
plt.show()

# Print transformation summary
print("Transformation Summary:")
print("=" * 40)
print(f"Initial direction ratios: 1 : 1 : 1")
print(f"Final direction ratios: {deformed_direction[0]:.3f} : {deformed_direction[1]:.3f} : {deformed_direction[2]:.3f}")
print(f"Stretch ratio: {np.linalg.norm(deformed_direction):.3f}")

# Calculate angle change
initial_unit = initial_direction / np.linalg.norm(initial_direction)
deformed_unit = deformed_direction / np.linalg.norm(deformed_direction)
angle_change = np.arccos(np.dot(initial_unit, deformed_unit)) * 180 / np.pi
print(f"Angle change: {angle_change:.1f}°")
```

### Summary of Results

**Part (a):** The line element with initial direction ratios 1:1:1 transforms to direction ratios:

**Part (b):** The sketch shows:

- There's significant rotation and stretching of the material line element

- The angular deviation from the original direction is about 26.6°

**Physical interpretation:** This deformation involves a combination of stretching, shearing, and rotation that significantly reorients material line elements, with the strongest component now in the x₁ direction.

## Problem 3

Find the components of the tensors F, C, B, F⁻¹, C⁻¹, B⁻¹, γ and κ for the deformation

$$x_1 = a_1(X_1 + \alpha X_2)$$
$$x_2 = a_2 X_2$$
$$x_3 = a_3 X_3$$

where, a₁, a₂, a₃ and α are constants. Find the conditions on these constants for the deformation to be possible in an incompressible material. A body which in the reference configuration is a unit cube with its edges parallel to the coordinate axes undergoes this deformation. Determine the length of its edges, and the angles between the edges, after the deformation. Sketch the deformed body.

## Solution to Problem 3

### Step 1: Calculate the Deformation Gradient F

From the given deformation mapping:
$$F_{ij} = \frac{\partial x_i}{\partial X_j}$$

$$\mathbf{F} = \begin{bmatrix}
a_1 & a_1\alpha & 0 \\
0 & a_2 & 0 \\
0 & 0 & a_3
\end{bmatrix}$$

### Step 2: Calculate F⁻¹

$$\mathbf{F}^{-1} = \begin{bmatrix}
\frac{1}{a_1} & -\frac{\alpha}{a_2} & 0 \\
0 & \frac{1}{a_2} & 0 \\
0 & 0 & \frac{1}{a_3}
\end{bmatrix}$$

### Step 3: Calculate Right Cauchy-Green Tensor C

$$\mathbf{C} = \mathbf{F}^T \mathbf{F} = \begin{bmatrix}
a_1^2 & a_1a_2\alpha & 0 \\
a_1a_2\alpha & a_1^2\alpha^2 + a_2^2 & 0 \\
0 & 0 & a_3^2
\end{bmatrix}$$

### Step 4: Calculate Left Cauchy-Green Tensor B

$$\mathbf{B} = \mathbf{F} \mathbf{F}^T = \begin{bmatrix}
a_1^2(1 + \alpha^2) & a_1a_2\alpha & 0 \\
a_1a_2\alpha & a_2^2 & 0 \\
0 & 0 & a_3^2
\end{bmatrix}$$

### Step 5: Calculate C⁻¹ and B⁻¹

$$\mathbf{C}^{-1} = \begin{bmatrix}
\frac{a_2^2 + a_1^2\alpha^2}{a_1^2a_2^2} & -\frac{\alpha}{a_2} & 0 \\
-\frac{\alpha}{a_2} & \frac{1}{a_2^2} & 0 \\
0 & 0 & \frac{1}{a_3^2}
\end{bmatrix}$$

$$\mathbf{B}^{-1} = \begin{bmatrix}
\frac{1}{a_1^2} & -\frac{\alpha}{a_1^2(1 + \alpha^2)} & 0 \\
-\frac{\alpha}{a_1^2(1 + \alpha^2)} & \frac{1 + \alpha^2}{a_2^2(1 + \alpha^2)} & 0 \\
0 & 0 & \frac{1}{a_3^2}
\end{bmatrix}$$

```python
# Numerical calculation for Problem 3
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define symbolic calculation first, then use specific values for visualization
def calculate_tensors(a1, a2, a3, alpha):
    """Calculate all deformation tensors for given parameters"""
    
    # Deformation gradient F
    F = np.array([
        [a1, a1*alpha, 0],
        [0, a2, 0],
        [0, 0, a3]
    ])
    
    # F inverse
    F_inv = np.array([
        [1/a1, -alpha/a2, 0],
        [0, 1/a2, 0],
        [0, 0, 1/a3]
    ])
    
    # Right Cauchy-Green tensor C
    C = F.T @ F
    
    # Left Cauchy-Green tensor B
    B = F @ F.T
    
    # Inverses
    C_inv = np.linalg.inv(C)
    B_inv = np.linalg.inv(B)
    
    # Green-Lagrangian strain tensor γ (or E)
    I = np.eye(3)
    gamma = 0.5 * (C - I)
    
    # Almansi strain tensor κ (or e)
    kappa = 0.5 * (I - B_inv)
    
    # Determinant of F (for incompressibility check)
    det_F = np.linalg.det(F)
    
    return F, F_inv, C, B, C_inv, B_inv, gamma, kappa, det_F

# Example with specific values
a1, a2, a3, alpha = 1.2, 0.8, 1.5, 0.3

F, F_inv, C, B, C_inv, B_inv, gamma, kappa, det_F = calculate_tensors(a1, a2, a3, alpha)

print("Deformation Gradient F:")
print(F)
print(f"\nDeterminant of F: {det_F:.4f}")
print(f"For incompressibility, det(F) = 1, so a1*a2*a3 = {1/det_F:.4f}")

print("\nF⁻¹:")
print(F_inv)

print("\nRight Cauchy-Green tensor C:")
print(C)

print("\nLeft Cauchy-Green tensor B:")
print(B)

print("\nGreen-Lagrangian strain tensor γ:")
print(gamma)

print("\nAlmansi strain tensor κ:")
print(kappa)
```

### Step 6: Green-Lagrangian and Almansi Strain Tensors

**Green-Lagrangian strain tensor γ (or E):**
$$\boldsymbol{\gamma} = \frac{1}{2}(\mathbf{C} - \mathbf{I}) = \begin{bmatrix}
\frac{a_1^2 - 1}{2} & \frac{a_1a_2\alpha}{2} & 0 \\
\frac{a_1a_2\alpha}{2} & \frac{a_1^2\alpha^2 + a_2^2 - 1}{2} & 0 \\
0 & 0 & \frac{a_3^2 - 1}{2}
\end{bmatrix}$$

**Almansi strain tensor κ (or e):**
$$\boldsymbol{\kappa} = \frac{1}{2}(\mathbf{I} - \mathbf{B}^{-1}) = \begin{bmatrix}
\frac{a_1^2 - 1}{2a_1^2} & \frac{a_1a_2\alpha}{2a_1^2(1 + \alpha^2)} & 0 \\
\frac{a_1a_2\alpha}{2a_1^2(1 + \alpha^2)} & \frac{a_2^2(1 + \alpha^2) - 1}{2a_2^2(1 + \alpha^2)} & 0 \\
0 & 0 & \frac{a_3^2 - 1}{2a_3^2}
\end{bmatrix}$$

### Step 7: Incompressibility Condition

For an incompressible material, the volume must be preserved:
$$\det(\mathbf{F}) = 1$$

$$\det(\mathbf{F}) = a_1 \cdot a_2 \cdot a_3 = 1$$

**Therefore: $a_1 a_2 a_3 = 1$**

### Step 8: Deformation of Unit Cube

**Initial cube edges:** Three unit vectors along coordinate axes
- $\mathbf{E}_1 = [1, 0, 0]^T$
- $\mathbf{E}_2 = [0, 1, 0]^T$  
- $\mathbf{E}_3 = [0, 0, 1]^T$

**Deformed edges:** $\mathbf{e}_i = \mathbf{F} \mathbf{E}_i$
- $\mathbf{e}_1 = [a_1, 0, 0]^T$ with length $|a_1|$
- $\mathbf{e}_2 = [a_1\alpha, a_2, 0]^T$ with length $\sqrt{a_1^2\alpha^2 + a_2^2}$
- $\mathbf{e}_3 = [0, 0, a_3]^T$ with length $|a_3|$

**Angles between deformed edges:**
- Angle between $\mathbf{e}_1$ and $\mathbf{e}_2$: $\cos^{-1}\left(\frac{a_1^2\alpha}{|a_1|\sqrt{a_1^2\alpha^2 + a_2^2}}\right)$
- Angle between $\mathbf{e}_1$ and $\mathbf{e}_3$: $90°$ (remain perpendicular)
- Angle between $\mathbf{e}_2$ and $\mathbf{e}_3$: $90°$ (remain perpendicular)

```python
# Calculate deformed cube properties
def analyze_deformed_cube(a1, a2, a3, alpha):
    """Analyze the deformation of a unit cube"""
    
    F = np.array([
        [a1, a1*alpha, 0],
        [0, a2, 0],
        [0, 0, a3]
    ])
    
    # Original unit cube edges
    E1 = np.array([1, 0, 0])
    E2 = np.array([0, 1, 0])
    E3 = np.array([0, 0, 1])
    
    # Deformed edges
    e1 = F @ E1
    e2 = F @ E2
    e3 = F @ E3
    
    # Edge lengths
    len1 = np.linalg.norm(e1)
    len2 = np.linalg.norm(e2)
    len3 = np.linalg.norm(e3)
    
    # Angles between edges
    angle_12 = np.arccos(np.dot(e1, e2) / (len1 * len2)) * 180 / np.pi
    angle_13 = np.arccos(np.dot(e1, e3) / (len1 * len3)) * 180 / np.pi
    angle_23 = np.arccos(np.dot(e2, e3) / (len2 * len3)) * 180 / np.pi
    
    return e1, e2, e3, len1, len2, len3, angle_12, angle_13, angle_23

# Use the same parameters
e1, e2, e3, len1, len2, len3, angle_12, angle_13, angle_23 = analyze_deformed_cube(a1, a2, a3, alpha)

print("Deformed Cube Analysis:")
print("=" * 30)
print(f"Edge 1 vector: [{e1[0]:.3f}, {e1[1]:.3f}, {e1[2]:.3f}], Length: {len1:.3f}")
print(f"Edge 2 vector: [{e2[0]:.3f}, {e2[1]:.3f}, {e2[2]:.3f}], Length: {len2:.3f}")
print(f"Edge 3 vector: [{e3[0]:.3f}, {e3[1]:.3f}, {e3[2]:.3f}], Length: {len3:.3f}")
print()
print("Angles between edges:")
print(f"Angle between edge 1 and edge 2: {angle_12:.1f}°")
print(f"Angle between edge 1 and edge 3: {angle_13:.1f}°")
print(f"Angle between edge 2 and edge 3: {angle_23:.1f}°")

# Theoretical calculations
print("\nTheoretical edge lengths:")
print(f"Edge 1 length: |a₁| = {abs(a1):.3f}")
print(f"Edge 2 length: √(a₁²α² + a₂²) = {np.sqrt(a1**2 * alpha**2 + a2**2):.3f}")
print(f"Edge 3 length: |a₃| = {abs(a3):.3f}")
```

```python
# Sketch the deformed cube
fig = plt.figure(figsize=(15, 5))

def plot_cube(ax, edges, title, color='blue'):
    """Plot a cube given its edge vectors"""
    # Origin
    origin = np.array([0, 0, 0])
    
    # Plot edges from origin
    ax.quiver(origin[0], origin[1], origin[2], 
              edges[0][0], edges[0][1], edges[0][2],
              color=color, linewidth=2, label='Edge 1')
    ax.quiver(origin[0], origin[1], origin[2], 
              edges[1][0], edges[1][1], edges[1][2],
              color='red', linewidth=2, label='Edge 2')
    ax.quiver(origin[0], origin[1], origin[2], 
              edges[2][0], edges[2][1], edges[2][2],
              color='green', linewidth=2, label='Edge 3')
    
    # Plot cube faces (simplified)
    vertices = np.array([
        [0, 0, 0],
        edges[0],
        edges[1],
        edges[2],
        edges[0] + edges[1],
        edges[0] + edges[2],
        edges[1] + edges[2],
        edges[0] + edges[1] + edges[2]
    ])
    
    # Draw some edges of the cube
    cube_edges = [
        [0, 1], [0, 2], [0, 3],  # from origin
        [1, 4], [1, 5],           # from edge 1
        [2, 4], [2, 6],           # from edge 2
        [3, 5], [3, 6],           # from edge 3
        [4, 7], [5, 7], [6, 7]   # to opposite corner
    ]
    
    for edge in cube_edges:
        start, end = edge
        ax.plot([vertices[start][0], vertices[end][0]],
                [vertices[start][1], vertices[end][1]],
                [vertices[start][2], vertices[end][2]],
                'k--', alpha=0.5, linewidth=1)
    
    ax.set_title(title)
    ax.legend()

# Original cube
ax1 = fig.add_subplot(131, projection='3d')
original_edges = [np.array([1, 0, 0]), np.array([0, 1, 0]), np.array([0, 0, 1])]
plot_cube(ax1, original_edges, 'Original Unit Cube')
ax1.set_xlim(0, 1.5)
ax1.set_ylim(0, 1.5)
ax1.set_zlim(0, 1.5)

# Deformed cube
ax2 = fig.add_subplot(132, projection='3d')
deformed_edges = [e1, e2, e3]
plot_cube(ax2, deformed_edges, 'Deformed Cube')
max_range = max(np.max(np.abs(e1)), np.max(np.abs(e2)), np.max(np.abs(e3))) * 1.2
ax2.set_xlim(-0.2, max_range)
ax2.set_ylim(-0.2, max_range)
ax2.set_zlim(-0.2, max_range)

# Side view showing shear
ax3 = fig.add_subplot(133)
# Project onto x1-x2 plane
ax3.quiver(0, 0, e1[0], e1[1], color='blue', linewidth=2, scale_units='xy', 
           angles='xy', scale=1, label='Edge 1 projection')
ax3.quiver(0, 0, e2[0], e2[1], color='red', linewidth=2, scale_units='xy', 
           angles='xy', scale=1, label='Edge 2 projection')

# Draw the parallelogram
corners = np.array([[0, 0], [e1[0], e1[1]], [e1[0]+e2[0], e1[1]+e2[1]], [e2[0], e2[1]], [0, 0]])
ax3.plot(corners[:, 0], corners[:, 1], 'k--', alpha=0.5)

ax3.set_xlim(-0.2, max_range)
ax3.set_ylim(-0.2, max_range)
ax3.set_aspect('equal')
ax3.set_title('x₁-x₂ Plane Projection\n(showing shear)')
ax3.legend()
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nDeformation Summary:")
print(f"Parameters: a₁={a1}, a₂={a2}, a₃={a3}, α={alpha}")
print(f"Volume change: {det_F:.3f} (should be 1.0 for incompressible)")
print(f"This deformation involves:")
print(f"- Stretching/compression in all directions")
print(f"- Shearing in the $x_1$-$x_2$ plane (α ≠ 0)")
print(f"- The angle between edges 1 and 2 changes from 90° to {angle_12:.1f}°")
```

### Summary of Results

**Tensor Components:**
- **F**: Upper triangular matrix with diagonal elements $a_1, a_2, a_3$ and shear term $a_1\alpha$
- **C, B**: Symmetric matrices showing material and spatial metrics

**Incompressibility Condition:**

**Deformed Unit Cube:**

**Physical Interpretation:**

3. The material maintains its volume when $a_1a_2a_3 = 1$

The deformation is particularly useful for modeling materials under combined normal and shear loading conditions.

## Problem 4: Rotation and Strain Tensors

Consider the rotation:
$$x_1 = X_1 \cos{\alpha} - X_2 \sin{\alpha}$$
$$x_2 = X_1 \sin{\alpha} + X_2 \cos{\alpha}$$
$$x_3 = X_3$$

Show that the Green-Lagrangian strain tensor is zero, and that the small strain theory result in a non-zero tensor.

### Solution

### Step 1: Calculate the deformation gradient F

For the given rotation:
- $x_1 = X_1 \cos\alpha - X_2 \sin\alpha$
- $x_2 = X_1 \sin\alpha + X_2 \cos\alpha$  
- $x_3 = X_3$

The deformation gradient is:
$$\mathbf{F} = \frac{\partial \mathbf{x}}{\partial \mathbf{X}} = \begin{bmatrix}
\cos\alpha & -\sin\alpha & 0 \\
\sin\alpha & \cos\alpha & 0 \\
0 & 0 & 1
\end{bmatrix}$$

This is a proper orthogonal matrix representing a rotation about the $X_3$ axis.

### Step 2: Calculate the Right Cauchy-Green Tensor C

$$\mathbf{C} = \mathbf{F}^T \mathbf{F} = \begin{bmatrix}
\cos\alpha & \sin\alpha & 0 \\
-\sin\alpha & \cos\alpha & 0 \\
0 & 0 & 1
\end{bmatrix} \begin{bmatrix}
\cos\alpha & -\sin\alpha & 0 \\
\sin\alpha & \cos\alpha & 0 \\
0 & 0 & 1
\end{bmatrix}$$

$$= \begin{bmatrix}
\cos^2\alpha + \sin^2\alpha & -\cos\alpha\sin\alpha + \sin\alpha\cos\alpha & 0 \\
-\sin\alpha\cos\alpha + \cos\alpha\sin\alpha & \sin^2\alpha + \cos^2\alpha & 0 \\
0 & 0 & 1
\end{bmatrix} = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix} = \mathbf{I}$$

### Step 3: Calculate the Green-Lagrangian Strain Tensor

$$\mathbf{E} = \frac{1}{2}(\mathbf{C} - \mathbf{I}) = \frac{1}{2}(\mathbf{I} - \mathbf{I}) = \mathbf{0}$$

**Result: The Green-Lagrangian strain tensor is zero, confirming that pure rotation produces no strain in the finite deformation theory.**

### Step 4: Calculate the Small Strain Tensor

For small strain theory, we use:
$$\boldsymbol{\varepsilon} = \frac{1}{2}(\nabla \mathbf{u} + (\nabla \mathbf{u})^T)$$

where $\mathbf{u} = \mathbf{x} - \mathbf{X}$ is the displacement vector.

The displacement components are:
$$u_1 = x_1 - X_1 = X_1(\cos\alpha - 1) - X_2\sin\alpha$$
$$u_2 = x_2 - X_2 = X_1\sin\alpha + X_2(\cos\alpha - 1)$$
$$u_3 = x_3 - X_3 = 0$$

The displacement gradient is:
$$\nabla \mathbf{u} = \begin{bmatrix}
\frac{\partial u_1}{\partial X_1} & \frac{\partial u_1}{\partial X_2} & \frac{\partial u_1}{\partial X_3} \\
\frac{\partial u_2}{\partial X_1} & \frac{\partial u_2}{\partial X_2} & \frac{\partial u_2}{\partial X_3} \\
\frac{\partial u_3}{\partial X_1} & \frac{\partial u_3}{\partial X_2} & \frac{\partial u_3}{\partial X_3}
\end{bmatrix} = \begin{bmatrix}
\cos\alpha - 1 & -\sin\alpha & 0 \\
\sin\alpha & \cos\alpha - 1 & 0 \\
0 & 0 & 0
\end{bmatrix}$$

The small strain tensor is:
$$\boldsymbol{\varepsilon} = \frac{1}{2}(\nabla \mathbf{u} + (\nabla \mathbf{u})^T) = \frac{1}{2}\begin{bmatrix}
2(\cos\alpha - 1) & 0 & 0 \\
0 & 2(\cos\alpha - 1) & 0 \\
0 & 0 & 0
\end{bmatrix}$$

$$= \begin{bmatrix}
\cos\alpha - 1 & 0 & 0 \\
0 & \cos\alpha - 1 & 0 \\
0 & 0 & 0
\end{bmatrix}$$

**Result: The small strain tensor is non-zero unless $\alpha = 0$.**

### Step 5: Physical Interpretation

**For finite rotation angles:**
- **Green-Lagrangian strain = 0**: Correctly captures that rotation involves no stretching or distortion
- **Small strain ≠ 0**: Incorrectly predicts strain due to the linearization assumption

**For small angles ($\alpha \ll 1$):**
Using $\cos\alpha \approx 1 - \frac{\alpha^2}{2}$ for small $\alpha$:
$$\varepsilon_{11} = \varepsilon_{22} = \cos\alpha - 1 \approx -\frac{\alpha^2}{2}$$

The small strain theory gives erroneous results proportional to $\alpha^2$, which becomes negligible only for very small rotations.

**Conclusion:**
This demonstrates a fundamental limitation of small strain theory: it cannot properly handle finite rotations, even pure rigid body rotations that should produce zero strain.

## Problem 5: Motion with Parameter A

A motion is given by:
$$x_1 = X_1 - AX_2$$
$$x_2 = X_2 - AX_1$$  
$$x_3 = -AX_1 + AX_2 + X_3$$

### Question (a)
Determine the Green-Lagrangian, the Almansi, and the strains from the infinitesimal strain theory.

**Solution:**

### Step 1: Calculate the Deformation Gradient F

From the given motion:
$$\mathbf{F} = \frac{\partial \mathbf{x}}{\partial \mathbf{X}} = \begin{bmatrix}
\frac{\partial x_1}{\partial X_1} & \frac{\partial x_1}{\partial X_2} & \frac{\partial x_1}{\partial X_3} \\
\frac{\partial x_2}{\partial X_1} & \frac{\partial x_2}{\partial X_2} & \frac{\partial x_2}{\partial X_3} \\
\frac{\partial x_3}{\partial X_1} & \frac{\partial x_3}{\partial X_2} & \frac{\partial x_3}{\partial X_3}
\end{bmatrix} = \begin{bmatrix}
1 & -A & 0 \\
-A & 1 & 0 \\
-A & A & 1
\end{bmatrix}$$

### Step 2: Calculate the Right Cauchy-Green Tensor C

$$\mathbf{C} = \mathbf{F}^T \mathbf{F} = \begin{bmatrix}
1 & -A & -A \\
-A & 1 & A \\
0 & 0 & 1
\end{bmatrix} \begin{bmatrix}
1 & -A & 0 \\
-A & 1 & 0 \\
-A & A & 1
\end{bmatrix}$$

$$= \begin{bmatrix}
1 + A^2 + A^2 & -A + A + A^2 & -A \\
-A - A + A^2 & A^2 + 1 + A^2 & A \\
-A & A & 1
\end{bmatrix} = \begin{bmatrix}
1 + 2A^2 & A^2 & -A \\
A^2 & 1 + 2A^2 & A \\
-A & A & 1
\end{bmatrix}$$

### Step 3: Calculate the Green-Lagrangian Strain Tensor E

$$\mathbf{E} = \frac{1}{2}(\mathbf{C} - \mathbf{I}) = \frac{1}{2}\begin{bmatrix}
2A^2 & A^2 & -A \\
A^2 & 2A^2 & A \\
-A & A & 0
\end{bmatrix} = \begin{bmatrix}
A^2 & \frac{A^2}{2} & -\frac{A}{2} \\
\frac{A^2}{2} & A^2 & \frac{A}{2} \\
-\frac{A}{2} & \frac{A}{2} & 0
\end{bmatrix}$$

### Step 4: Calculate F⁻¹ for the Almansi Strain Tensor

$$\det(\mathbf{F}) = 1 \cdot (1 \cdot 1 - 0 \cdot A) - (-A) \cdot (-A \cdot 1 - 0 \cdot (-A)) + 0 = 1 - A^2$$

$$\mathbf{F}^{-1} = \frac{1}{1-A^2}\begin{bmatrix}
1 & A & 0 \\
A & 1 & 0 \\
A^2 & -A & 1-A^2
\end{bmatrix}$$

### Step 5: Calculate the Left Cauchy-Green Tensor B

$$\mathbf{B} = \mathbf{F} \mathbf{F}^T = \begin{bmatrix}
1 & -A & 0 \\
-A & 1 & 0 \\
-A & A & 1
\end{bmatrix} \begin{bmatrix}
1 & -A & -A \\
-A & 1 & A \\
0 & 0 & 1
\end{bmatrix}$$

$$= \begin{bmatrix}
1 + A^2 & -A - A & -A + A^2 \\
-A - A & A^2 + 1 & A + A \\
-A + A^2 & A + A & A^2 + A^2 + 1
\end{bmatrix} = \begin{bmatrix}
1 + A^2 & -2A & A^2 - A \\
-2A & 1 + A^2 & 2A \\
A^2 - A & 2A & 2A^2 + 1
\end{bmatrix}$$

### Step 6: Calculate the Almansi Strain Tensor e

$$\mathbf{e} = \frac{1}{2}(\mathbf{I} - \mathbf{B}^{-1})$$

This requires calculating $\mathbf{B}^{-1}$, which is complex. Alternatively, we can use:
$$\mathbf{e} = \frac{1}{2}(\mathbf{I} - (\mathbf{F}\mathbf{F}^T)^{-1}) = \frac{1}{2}(\mathbf{I} - (\mathbf{F}^{-1})^T\mathbf{F}^{-1})$$

### Step 7: Calculate the Infinitesimal Strain Tensor ε

The displacement vector is:
$$\mathbf{u} = \mathbf{x} - \mathbf{X} = \begin{bmatrix}
-AX_2 \\
-AX_1 \\
-AX_1 + AX_2
\end{bmatrix}$$

The displacement gradient is:
$$\nabla \mathbf{u} = \begin{bmatrix}
0 & -A & 0 \\
-A & 0 & 0 \\
-A & A & 0
\end{bmatrix}$$

The infinitesimal strain tensor is:
$$\boldsymbol{\varepsilon} = \frac{1}{2}(\nabla \mathbf{u} + (\nabla \mathbf{u})^T) = \frac{1}{2}\begin{bmatrix}
0 & -2A & -A \\
-2A & 0 & A \\
-A & A & 0
\end{bmatrix} = \begin{bmatrix}
0 & -A & -\frac{A}{2} \\
-A & 0 & \frac{A}{2} \\
-\frac{A}{2} & \frac{A}{2} & 0
\end{bmatrix}$$

```python
# Numerical solution for Problem 5a
import numpy as np
import sympy as sp

# Define symbolic parameter A
A = sp.Symbol('A', real=True)

print("Problem 5a: Motion Analysis")
print("="*50)

# Define deformation gradient F
F_sym = sp.Matrix([
    [1, -A, 0],
    [-A, 1, 0],
    [-A, A, 1]
])

print("Deformation gradient F:")
sp.pprint(F_sym)
print()

# Calculate determinant
det_F = F_sym.det()
print(f"det(F) = {det_F}")
print()

# Right Cauchy-Green tensor C
C_sym = F_sym.T * F_sym
print("Right Cauchy-Green tensor C:")
sp.pprint(C_sym)
print()

# Green-Lagrangian strain tensor E
I = sp.eye(3)
E_sym = sp.Rational(1,2) * (C_sym - I)
print("Green-Lagrangian strain tensor E:")
sp.pprint(E_sym)
print()

# Left Cauchy-Green tensor B
B_sym = F_sym * F_sym.T
print("Left Cauchy-Green tensor B:")
sp.pprint(B_sym)
print()

# Calculate F inverse
F_inv_sym = F_sym.inv()
print("F⁻¹:")
sp.pprint(F_inv_sym)
print()

# Almansi strain tensor e
# e = 1/2(I - B^(-1)) = 1/2(I - (F^(-1))^T * F^(-1))
FT_inv_F_inv = F_inv_sym.T * F_inv_sym
e_sym = sp.Rational(1,2) * (I - FT_inv_F_inv)
print("Almansi strain tensor e:")
sp.pprint(e_sym)
print()

# Displacement gradient
grad_u = sp.Matrix([
    [0, -A, 0],
    [-A, 0, 0],
    [-A, A, 0]
])

# Infinitesimal strain tensor
eps_sym = sp.Rational(1,2) * (grad_u + grad_u.T)
print("Infinitesimal strain tensor ε:")
sp.pprint(eps_sym)
print()

# Numerical example with A = 0.1
A_val = 0.1
print(f"\nNumerical example with A = {A_val}:")
print("-"*40)

F_num = np.array(F_sym.subs(A, A_val), dtype=float)
C_num = np.array(C_sym.subs(A, A_val), dtype=float)
E_num = np.array(E_sym.subs(A, A_val), dtype=float)
B_num = np.array(B_sym.subs(A, A_val), dtype=float)
e_num = np.array(e_sym.subs(A, A_val), dtype=float)
eps_num = np.array(eps_sym.subs(A, A_val), dtype=float)

print("Green-Lagrangian strain E:")
print(E_num)
print("\nAlmansi strain e:")
print(e_num)
print("\nInfinitesimal strain ε:")
print(eps_num)
```

Compare the results from the infinitesimal strain theory with the results obtained from the spatial and material strain tensors taking into account the infinitesimal strain hypotheses. What are your conclusions?

### Step 2: Analysis Under Small Deformation Assumptions

For small deformations, we assume $A \ll 1$. Let's expand the finite strain tensors:

**Green-Lagrangian strain E for small A:**
- Diagonal terms: $E_{11} = E_{22} = A^2 \approx 0$ (second-order)
- Off-diagonal terms: $E_{12} = E_{21} = \frac{A^2}{2} \approx 0$ (second-order)
- Terms: $E_{13} = E_{31} = -\frac{A}{2}$, $E_{23} = E_{32} = \frac{A}{2}$ (first-order)

$$\mathbf{e} \approx \frac{1}{2}(\mathbf{I} - (\mathbf{I} - 2\boldsymbol{\varepsilon})) = \boldsymbol{\varepsilon}$$

### Step 3: Key Differences and Similarities

**Similarities:**
1. All three tensors have the same pattern for first-order terms in A
2. The off-diagonal shear components involving the 3-direction are identical:
   - $E_{13} = \varepsilon_{13} = -\frac{A}{2}$
   - $E_{23} = \varepsilon_{23} = \frac{A}{2}$

**Key Differences:**
1. **Diagonal terms:**
   - Green-Lagrangian: $E_{11} = E_{22} = A^2$ (nonzero, second-order)

   - Green-Lagrangian: $E_{12} = \frac{A^2}{2}$ (second-order)
   - Infinitesimal: $\varepsilon_{12} = -A$ (first-order)

3. **Physical interpretation:**
   - The infinitesimal theory captures the dominant shear deformation in the 1-2 plane
   - The Green-Lagrangian theory includes geometric nonlinearities (second-order effects)

### Step 4: Convergence Analysis

For small A, the relationship between the tensors is:
$$\mathbf{E} = \boldsymbol{\varepsilon} + \text{higher-order terms}$$

Specifically:
$$\mathbf{E} = \boldsymbol{\varepsilon} + \frac{1}{2}(\nabla \mathbf{u})^T \nabla \mathbf{u}$$

where the second term represents geometric nonlinearity.

### Step 5: Physical Conclusions

1. **Accuracy for small A:** When $A \ll 1$, the infinitesimal strain theory provides a good first-order approximation, especially for the dominant shear terms.

2. **Geometric nonlinearity:** The differences arise from the geometric nonlinearity inherent in finite deformation. The Green-Lagrangian strain correctly accounts for the coupling between different deformation modes.

3. **Dominant deformation mode:** This motion represents a combination of:
   - Simple shear in the 1-2 plane (dominant, captured by all theories)
   - Shear in the 1-3 and 2-3 planes
   - Secondary normal strains due to geometric coupling (only in finite theory)

4. **Practical implications:** For engineering applications where $A < 0.1$, the infinitesimal theory gives reasonable results for most components, but finite strain theory is essential when geometric nonlinearity matters.

Draw a diagram for $\varepsilon_{11}$ and $E_{11}$ as a function of $\lambda = l/L$, $0 < \lambda < 3$ for $\theta = 0$

```python
import numpy as np

import matplotlib.pyplot as plt

# Range of lambda values
lambda_vals = np.linspace(0.1, 3, 100)  # avoid zero to prevent division by zero

# For theta = 0
theta = 0

# Green-Lagrange strain E_11 as a function of lambda
E_11 = 0.5 * (lambda_vals**2 - 1)

# Small strain theory: epsilon_11 = lambda - 1 (for small strains, but plotted for all lambda)
epsilon_11 = lambda_vals - 1

plt.figure(figsize=(8, 5))
plt.plot(lambda_vals, E_11, label=r'Green-Lagrange $E_{11}$', color='blue')
plt.plot(lambda_vals, epsilon_11, label=r'Small strain $\varepsilon_{11}$', color='red', linestyle='--')
plt.axhline(0, color='gray', linewidth=0.7, linestyle=':')
plt.xlabel(r'$\lambda = l/L$')
plt.ylabel(r'Strain')
plt.title(r'$\varepsilon_{11}$ and $E_{11}$ vs. $\lambda$ for $\theta=0$')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Question (f)
Draw the corresponding diagram as a function of $\theta$ for $0 < \theta < \pi/2$ when $l = 1.01L$.

**Solution:**

```python
import numpy as np

import matplotlib.pyplot as plt

# Given: l = 1.01L, so lambda = 1.01
lambda_fixed = 1.01

# Range of theta values from 0 to pi/2
theta_vals = np.linspace(0, np.pi/2, 200)

# Green-Lagrange strain E_11 as a function of theta
E_11_theta = 0.5 * (lambda_fixed**2 * np.cos(theta_vals)**2 + (1/lambda_fixed) * np.sin(theta_vals)**2 - 1)

# Small strain theory: epsilon_11 = lambda*cos(theta) - 1
epsilon_11_theta = lambda_fixed * np.cos(theta_vals) - 1

plt.figure(figsize=(8, 5))
plt.plot(theta_vals, E_11_theta, label=r'Green-Lagrange $E_{11}$', color='blue')
plt.plot(theta_vals, epsilon_11_theta, label=r'Small strain $\varepsilon_{11}$', color='red', linestyle='--')
plt.axhline(0, color='gray', linewidth=0.7, linestyle=':')
plt.xlabel(r'$\theta$ (radians)')
plt.ylabel(r'Strain')
plt.title(r'$\varepsilon_{11}$ and $E_{11}$ vs. $\theta$ for $\lambda=1.01$')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
